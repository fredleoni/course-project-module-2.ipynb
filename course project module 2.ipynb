```python
#check the ratio per countries, by adding countries column to previous dataframe
macro2=food[['product_name','fat_100g','carbohydrates_100g','proteins_100g','countries']].copy()
macro2.head()
```

```python
#take of the en: in front of countries
country=macro2.countries

country[country.str.contains('en:', na=False)]
country.head()
```

```python
#strip the en: part, and only keep the first country
def  transform(s):
    s=s.str.replace('en:','')
    s=s.str.replace('BE','Belgium')
    s=s.str.replace('Belgique','Belgium')
    s=s.str.replace('België','Belgium')
    s=s.str.replace('Frankreich','France')
    s=s.str.replace('FR','France')
    s=s.str.replace('GB','UK')
    s=s.str.replace('México','Mexico')
    s=s.str.replace('États-Unis','United States')
    s=s.str.replace('USA','United States')
    s=s.str.replace('US','United States')
   

    return(s)

country= transform(macro2.countries)
```

```python
#separate when they are different countries
country_df=country.str.split(',', expand=True)
country_df.head()
```

```python
#only keep the first country
macro2.countries=country_df[0]
macro2.head()
```

```python
#as only fat, carbo and proteins contains calories we need to select the row where there is no 0 in the columns


macro3=macro2[(macro2['fat_100g'] !=0) & (macro2['carbohydrates_100g']!=0)& (macro2['proteins_100g']!=0)]
macro3.head()
```

```python
#check some random cell in countries column
macro3.countries[25000]
```

```python
#group by countries and make the mean
mac=macro3.groupby('countries')['fat_100g','carbohydrates_100g','proteins_100g'].mean()

mac
```

```python
#Macronutrient ratio per country

mac.loc[:,['fat_100g','carbohydrates_100g','proteins_100g']].plot.barh(stacked=True,  figsize=(10,20))
plt.ylabel('Country')
plt.xlabel('Macronutrient percentage')
plt.title ('Macronutrient ratio per country')
```
